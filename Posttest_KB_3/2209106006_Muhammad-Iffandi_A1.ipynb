# Distribusi Pegulat Berdarkan asal Federasinya

```python
plt.hist(df['federation'], bins=20, color='blue', edgecolor='black')
plt.xlabel('FEDERASI')
plt.ylabel('Jumlah Pegulat')
plt.title('Distribusi Pegulat Berdasarkan asal Federasi')
plt.xticks(rotation = 30, ha='right')
plt.show()
```

## Kesimpulan
1. Fight Zone Wrestling (FZW): Terdapat sekitar 200 pegulat yang terkait dengan Fight Zone Wrestling. Ini menunjukkan bahwa FZW memiliki jumlah pegulat yang signifikan dalam dataset ini.
2. World Wrestling Entertainment (WWE): Terdapat sekitar 100 pegulat yang terkait dengan WWE. Meskipun jumlah ini lebih rendah dibandingkan dengan FZW, WWE masih memiliki jumlah pegulat yang cukup besar.
3. Consejo Global de Lucha Libre (CMLL): Terdapat sekitar 180 pegulat yang terkait dengan Consejo Global de Lucha Libre. Jumlah ini berada di antara FZW dan WWE, menunjukkan bahwa CMLL juga memiliki jumlah pegulat yang cukup signifikan.

Kesimpulannya, FZW memiliki jumlah pegulat terbanyak dalam dataset, diikuti oleh CMLL dan kemudian WWE.

# Persentase perbandingan jumlah pegulat berdasarkan jenis kelamin

```python
gender_count = df['gender'].value_counts()
plt.pie(gender_count, labels = gender_count.index, autopct = '%1.0f%%' , startangle=85)
plt.title('Persentase Berdasarkan Jenis Kelamin')
plt.show()
```

## Kesimpulan
1. Grafik pie menunjukkan bahwa sekitar 53% dari pegulat dalam dataset tersebut adalah laki-laki, sedangkan sekitar 47% sisanya adalah perempuan.
2. Laki-laki mendominasi jumlah pegulat dalam dataset ini dengan persentase yang lebih tinggi. Ini bisa mencerminkan pola umum dalam gulat profesional di mana laki-laki sering kali mendominasi lebih banyak dalam jumlah pegulat dibandingkan perempuan.
3. Meskipun laki-laki mendominasi dalam jumlah, penting untuk diingat bahwa keragaman gender dalam gulat profesional semakin diakui dan dihargai. Ada banyak pegulat perempuan yang sangat berbakat dan sukses dalam industri gulat, dan peran mereka semakin terlihat.

kita dapat menyimpulkan bahwa dalam dataset tersebut, laki-laki memiliki jumlah pegulat yang lebih tinggi dibandingkan perempuan, tetapi keragaman gender tetap menjadi hal yang penting dalam dunia gulat profesional.

# Jumlah Pegulat berdasarkan Asal Negaranya

```python
nationality_counts = df['nationality'].value_counts()

plt.figure(figsize=(12, 6))
nationality_counts.plot(kind='bar', color='RED')
plt.title('Jumalah Pegulat Berdasarkan Asal Negara')
plt.xlabel('Negara Kebangsaan')
plt.ylabel('Jumlah Pegulat')
plt.xticks(rotation=45)
plt.show()
```

3. Meskipun France, Japan, dan USA mendominasi dalam jumlah pegulat, ada juga perwakilan dari negara-negara lain seperti Australia, Egypt, England, Argentina, dan India. Ini mencerminkan keragaman peserta dari berbagai negara di turnamen gulat ini.
4. Beberapa negara seperti India dan Argentina memiliki jumlah peserta yang lebih rendah dibandingkan dengan negara-negara lainnya. Ini mungkin menunjukkan bahwa gulat belum begitu populer atau dikembangkan di negara-negara tersebut dalam konteks turnamen ini.
5. Data ini juga bisa digunakan untuk mengidentifikasi negara-negara yang memiliki potensi untuk mendominasi dunia gulat di masa depan. Misalnya, India dengan hanya 10 pegulat bisa saja mengembangkan lebih banyak atlet gulat yang kompetitif di masa mendatang.a 

# Korelasi antar Atribute Numerik

```python
Korelasi = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(Korelasi, annot=True)
```

## Kesimpulan
Terlihat hubungan/korelasi antar atribut terdapat positif dan negatif. contoh:
1. korelasi positif antara height dan weight : Dalam konteks ini, semakin tinggi tinggi badan (height), semakin besar berat badan (weight). Dengan kata lain, orang yang lebih tinggi cenderung memiliki berat badan yang lebih besar.
2. korelasi negatif antara year_start dan age: semakin tua seorang pegulat (nilai "age" yang lebih tinggi), kemungkinan besar dia akan memulai karir gulatnya pada tahun yang lebih awal (nilai "year_start" yang lebih rendah).
