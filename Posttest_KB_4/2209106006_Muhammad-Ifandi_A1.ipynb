## 4. Melakukan Data Cleaning pada data dengan nilai null

```python
df.isnull().sum()
```

### 1. isnull() -> berfungsi untuk menampilkan kolom dengan nilai null

### 2. sum() -> Berfungsi untuk menghitung jumlah elemen dalam suatu rangkaian data

```python
df[['metacritic']]= df[['metacritic']].fillna(df['metacritic'].mean())
df[['released']]= df[['released']].fillna(df['released'].mode()[0])
df[['tags']]= df[['tags']].fillna(df['tags'].mode()[0])
df[['genres']]= df[['genres']].fillna(df['genres'].mode()[0])
```

### fillna() -> Adalah metode dalam library Pandas yang digunakan untuk mengisi nilai-nilai yang hilang (NaN, Not-a-Number) dalam DataFrame atau Series dengan nilai tertentu

### mean() -> Digunakan untuk mengisi nilai-nilai yang hilang (NaN) dalam suatu kolom dalam DataFrame dengan nilai rata-rata dari kolom tersebut

### mode() -> Digunakan untuk mengisi nilai-nilai yang hilang (NaN) dalam suatu DataFrame dengan modus (mode) dari kolom tersebut

```python
df.isnull().sum()
```
